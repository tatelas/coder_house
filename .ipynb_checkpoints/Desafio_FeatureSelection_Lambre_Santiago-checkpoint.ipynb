```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans

from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import scale
import statsmodels.api as sm
```

```python
df_data = pd.read_excel("_ml_jug.xlsx")
```

```python
round(df_data.describe(),2)
```

Aprendizaje no supervisado con Kmeans

```python
model = KMeans(n_clusters=3, random_state= 43)
```

```python
cc = pd.DataFrame()
cc = cc.assign(name = df_data["nombre"])
cc = cc.assign(pxg = df_data["pts_ganados"]/df_data['games_jug'])
cc = cc.assign(nfxg = df_data["NF_total"]/df_data['games_jug'])
cc = cc.assign(rem = (df_data["sm_total"] + df_data["nsmtotal"])/df_data['games_jug'])

cc = cc.groupby("name")[["pxg","nfxg","rem"]].mean().sort_values(by="pxg", ascending=False)

c_index = cc.index
cc.reset_index(inplace =True, drop=True)

#df 
cc1 = cc

#df.values np
cc = cc.values

cc1.head(2)
```

Metodo del codo par obtener numero de clusters

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(cc)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.scatter(range(1, 11), wcss)
plt.title('Metodo del codo')
plt.xlabel('Numero de clusters')
plt.ylabel('Inercia')
```

Si bien las posibilidades para agrupar podrian ser desde 2 cluster hasta 4, el an√°lisis previo del deporte divide a los jugadores en 3 grupos: Los que se destacan por la cantidad de puntos que hacen(ofensivos), los que se destacan defensivamente (defensivos), y en el medio jugadores hibridos, asi que elegimos separar a los jugadores en 3 clusters

Graficar centromeros

```python
#3 Centromeros 
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(cc)
plt.scatter(cc[:,0], cc[:,1])
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(cc)
plt.scatter(cc[:,0], cc[:,1], label = "No forzados")
plt.scatter(cc[:,0], cc[:,2], label ="Remates", color="black",alpha=0.7)
plt.xlabel("Puntos por game")
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red')
plt.legend(loc='best')
plt.show()
```

Se puede ver en el grafico que en cuanto a los remates los 3 centromeros separa en baja, intermedia y alta cantidad de remates segun los puntos por game.
En cuanto a los errores no forzados la separaci√≥n no sencilla de explicar visualmente

Se realiza un pipeline para escalar datos y realizar PCA

```python
cc =cc1
# Entrenamiento modelo PCA con escalado de los datos
pca_pipe = make_pipeline(StandardScaler(), PCA())
pca_pipe.fit(cc)

# Se extrae el modelo entrenado del pipeline
modelo_pca = pca_pipe.named_steps['pca']


# Porcentaje de varianza explicada por cada componente
print('Porcentaje de varianza explicada por cada componente')
print(modelo_pca.explained_variance_ratio_)
import seaborn as sns;sns.set_style("whitegrid")
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))
ax.bar(x= np.arange(modelo_pca.n_components_) + 1,height = modelo_pca.explained_variance_ratio_)
for x, y in zip(np.arange(len(cc.columns)) + 1, modelo_pca.explained_variance_ratio_):
    label = round(y, 2)
    ax.annotate(label,(x,y),textcoords="offset points",xytext=(0,10),ha='center')
ax.set_xticks(np.arange(modelo_pca.n_components_) + 1);ax.set_ylim(0, 1.1)
ax.set_title('Porcentaje de varianza explicada por cada componente')
ax.set_xlabel('Componente principal')
ax.set_ylabel('Por. varianza explicada')
```

Teniendo en cuenta que son tan solo 3 componentes, no parece necesario disminuirlo a 2, que explicarian 86,5% de la varianza

Quedarnos solo con 2 componentes

```python
num_components_to_keep = 2
pca_pipe = make_pipeline(StandardScaler(), PCA(n_components=num_components_to_keep))
pca_pipe.fit(cc)

# Se extraen los 2 primeros componentes principales
modelo_pca_2_components = pca_pipe.named_steps['pca']

# Porcentaje de varianza explicada por cada componente
print('Porcentaje de varianza explicada por cada componente')
print(modelo_pca_2_components.explained_variance_ratio_)
```

Evaluacion del modelo si a los puntos que hace por game un jugador, le restamos los errores no forzados que comente en ese game, quedando en esa nueva variable los puntos netos que aporta cada jugador al game

```python
cc1["puntos_netos"] = round(cc1.pxg - cc1.nfxg,3)
df_sintetico = cc1

df_sintetico.head(2)
```

Quitamos pxg y nfxg de modelo

```python
df_sintetico = df_sintetico[["rem", "puntos_netos"]]
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(df_sintetico)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.scatter(range(1, 11), wcss)
plt.title('Metodo del codo')
plt.xlabel('Numero de clusters')
plt.ylabel('Inercia')
```

Despues del cluster 3 comienza a disminuir mas lentamente la curva. Seleccionamos el 3 como el numero de cluster

```python

# Entrenamiento modelo PCA con escalado de los datos
pca_pipe = make_pipeline(StandardScaler(), PCA(n_components=2))
pca_pipe.fit(df_sintetico)

# Se extrae el modelo entrenado del pipeline
modelo_pca = pca_pipe.named_steps['pca']


# Porcentaje de varianza explicada por cada componente
print('Porcentaje de varianza explicada por cada componente')
print(modelo_pca.explained_variance_ratio_*100)

import seaborn as sns;sns.set_style("whitegrid")
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 5))
ax.bar(x= np.arange(modelo_pca.n_components_) + 1,height = modelo_pca.explained_variance_ratio_)
for x, y in zip(np.arange(len(df_sintetico.columns)) + 1, modelo_pca.explained_variance_ratio_):
    label = round(y, 2)
    ax.annotate(label,(x,y),textcoords="offset points",xytext=(0,10),ha='center')
ax.set_xticks(np.arange(modelo_pca.n_components_) + 1);ax.set_ylim(0, 1.1)
ax.set_title('Porcentaje de varianza explicada por cada componente')
ax.set_xlabel('Componente principal')
ax.set_ylabel('Por. varianza explicada')
```

Hubo un notorio cambio en el componente 1 al juntar 2 caracteristicas, que ahora explica el 74% de la varianza, cuando antes explicaba el 56%. Entre 2 componentes explican el 100%, y cuando seleccionamos 3 variables los 2 mayores componentes explicaban el 87%

```python
def g_bar(df):
    # Crea el gr√°fico scatter con nombres
    df= df.sort_values("cluster_label", ascending=False)
    
    fig, axs = plt.subplots(2,1,figsize=(7,5))
    fig.set_facecolor(fondo)
    name1 = "pts netos"
    name2 = "Rem"
    
    ax=axs[0]
    width = 0.2
    index = np.arange(3)
    
    
    ax.patch.set_facecolor(fondo)
    ax.scatter(df["cluster_label"].sort_values() , "puntos_netos", data=df,label='pts netos',color=celeste)
    
    ax.scatter(df["cluster_label"].sort_values()  + 0.1, "rem", data=df,label="Remate",color="yellow")
    
    ax.set_ylabel("Cantidad por game", color="w")
    ax.set_xlabel("Labels - cluster",color="w")
    ax.set_xticks(df["cluster_label"] + width, labels)
    plt.margins(0.02)
    ax.text(-0.1,1.5,f"{name1}-", color=celeste, fontsize=11)
    ax.text(0.16,1.5,name2, color="yellow", fontsize=11)
    

    ax.tick_params(axis="y", colors = "w")
    ax.tick_params(axis="x", colors = "w")
    ax.set_title("Clusterizaci√≥n", color ="w")
    ax.xaxis.grid(False)  # Desactivar las l√≠neas de la cuadr√≠cula en el eje x
    ax.yaxis.grid(False)
    ax.grid(False)
    
    
    ax=axs[1]
    
    ax.patch.set_facecolor(fondo)
    #sns.barplot(x= df.index, y=df["rem"], hue=df["cluster_label"])
    # Define una paleta de colores personalizada
    colores = [celeste, verde_lindo, "red"]

    # Asigna colores a las barras en funci√≥n de los valores de "cluster_label"
    colores_barras = [colores[label] for label in df["cluster_label"]]
    ax.bar(df.index, df["puntos_netos"], color=colores_barras)
    ax.scatter(df.index, df["rem"], color ="yellow", s=8)
    plt.xticks(rotation=90)
    ax.set_ylabel("Puntos por game y remates", color = "w")
    

    #cambiar colores
    ax.tick_params(axis="y", colors="white")
    
    ax.tick_params(axis="x", colors="white")
   
    
    plt.title("Clusterizaci√≥n", color="w")
    ax.xaxis.grid(False)  # Desactivar las l√≠neas de la cuadr√≠cula en el eje x
    ax.yaxis.grid(False)
    ax.grid(False)
    
    ax.text(0.25, 1.5, "Remates", color="yellow", weight='bold', ha='center')
    
    # Mostrar el gr√°fico
    plt.tight_layout()
    plt.show()
g_bar(df_sintetico)
```

La penalizacion por los errorez no forzados a la cantidad de puntos por game, conllevo a que jugadores que hacian muchos puntos y estaban en el cluster azul (mas puntos por game hacen) cambien y pasen al cluster verde, que es el grupo intermedio.
La cantidad de remates contin√∫a siendo un factor importante para separar a los jugadores, y los que mas puntos hacen, son los que rematan mas la gran mayoria de los casos
